# RMSE BAR PLOT

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
d = {'Boost-LR': [0.037,0.034,0.119,3.22], 'XgBoost': [0.27,0.396,1.45,3.74],'CatBoost': [0.052,0.045,1.44,5.23]}
data=pd.DataFrame(data=d, index=["Bruska data","Jelinak data", "Inland data", "Texas data"])
print(data)
x1=data['Boost-LR']
x2=data['XgBoost']
x3=data['CatBoost']
#x4=data['LSTM']
y=["Bruska data","Jelinak data", "Inland data", "Texas data"]
sns.set_style("white")
y_axis=np.arange(len(y))
hatches = ['*', '.', 'x', '/']
plt.ylabel("RMSE (kW)")
plt.rcParams['font.size'] = '14'
plt.rcParams["font.family"] = "Times New Roman"
plt.legend(prop={"size":12})
plt.rcParams["figure.figsize"] = (8,6)
plt.bar(y_axis, x1,0.2, label= 'Boost-LR' ,edgecolor='black', color=['silver'],hatch=hatches[0])
plt.bar(y_axis+0.2, x2,0.2, label= 'XgBoost' ,edgecolor='black', color=['royalblue'],hatch=hatches[1])
plt.bar(y_axis+0.4, x3,0.2, label= 'CatBoost',edgecolor='black',color=['sandybrown'],hatch=hatches[2] )
#plt.bar(y_axis+0.6, x4,0.2, label= 'LSTM',edgecolor='black', color=['darkseagreen'],hatch=hatches[3] )
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.legend()
plt.savefig('RMSE_barplot.pdf') 
plt.savefig('RMSE_barplot.eps') 
plt.savefig('RMSE_barplot.png') 

```

# MAE BAR PLOT

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

d = {'Boost-LR': [0.024,0.019,0.071,2.23], 'XgBoost': [0.095,0.181,1.03,2.52],'CatBoost': [0.035,0.028,0.098,3.66]}
data=pd.DataFrame(data=d, index=["Bruska data","Jelinak data", "Inland data", "Texas data"])
print(data)
x1=data['Boost-LR']
x2=data['XgBoost']
x3=data['CatBoost']
y=["Bruska data","Jelinak data", "Inland data", "Texas data"]
sns.set_style("white")
y_axis=np.arange(len(y))
hatches = ['*', '.', 'x', '/']
plt.ylabel("MAE (kW)")
plt.rcParams['font.size'] = '14'
plt.rcParams["font.family"] = "Times New Roman"
plt.legend(prop={"size":12})
plt.rcParams["figure.figsize"] = (8,6)
plt.bar(y_axis, x1,0.2, label= 'Boost-LR' ,edgecolor='black', color=['silver'],hatch=hatches[0])
plt.bar(y_axis+0.2, x2,0.2, label= 'XgBoost' ,edgecolor='black', color=['royalblue'],hatch=hatches[1])
plt.bar(y_axis+0.4, x3,0.2, label= 'CatBoost',edgecolor='black',color=['sandybrown'],hatch=hatches[2] )
plt.xticks(y_axis, y)
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
#plt.xlabel("Years")
plt.legend() 
plt.savefig('MAE_barplot.pdf') 
plt.savefig('MAE_barplot.eps') 
plt.savefig('MAE_barplot.png') 
```
