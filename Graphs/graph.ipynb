# TEXAS GRAPHICALL REPRESENTATION

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Load the data
df= pd.read_csv("Hybrid_output_for_Texas1.csv")
features=list(df)

print(features)
df= df.head(100)
# Set up the main plot
plt.rcParams['font.size'] = '16'
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["figure.figsize"] = (10, 8)
plt.plot(df['Actual'], label= 'Measured')
plt.plot(df['Boost-LR'],label='Boost-LR')
plt.plot(df['CatBoost'], label= 'CatBoost')
plt.plot(df['RF'],label='RF' )
plt.plot(df['XgBoost'], label= 'XgBoost')
plt.plot(df['Transformer'],label='Transformer')
plt.plot(df['Informer'],label='Informer' )
plt.grid(color = 'black', linestyle = '--', linewidth = 0.2)
plt.xlabel('Time (h)',fontweight='bold')
plt.ylabel("Wind power (kW)", fontweight='bold')
plt.legend(prop={"size":14} )

# Identify the points for the zoomed-in section
zoom_start = 8
zoom_end = 11

# Define the position and size of the inset
axins = inset_axes(plt.gca(), width="50%", height="30%", loc='upper center')

# Plot the zoomed-in data in the inset
axins.plot(df['Actual'][zoom_start:zoom_end + 1],linestyle='--', label='Measured')
axins.plot(df['Boost-LR'][zoom_start:zoom_end + 1], label='NeuroSpike',linestyle='--')
axins.plot(df['LSTM'][zoom_start:zoom_end + 1], label='LSTM',linestyle='--',)
axins.plot(df['GRU'][zoom_start:zoom_end + 1], label='GRU',linestyle='--')
axins.plot(df['CatBoost'][zoom_start:zoom_end + 1], label='CatBoost',linestyle='--')
axins.plot(df['XgBoost'][zoom_start:zoom_end + 1], label='XgBoost',linestyle='--')
axins.plot(df['RF'][zoom_start:zoom_end + 1], label='RF',linestyle='--')
axins.plot(df['Transformer'][zoom_start:zoom_end + 1], label='Transformer',linestyle='--')
axins.plot(df['Informer'][zoom_start:zoom_end + 1], label='Informer',linestyle='--')

# Customize and annotate the inset
#axins.set_xlabel('Time (h)')
#axins.set_ylabel('GHI (watt per meter square)')
#axins.set_title('Zoomed-In View')
#axins.legend(prop={"size": 8})
axins.tick_params(axis='x', labelsize=10)  # Adjust the font size as needed
axins.tick_params(axis='y', labelsize=10)
# Connect the main graph and inset with lines or arrows
plt.annotate("",(zoom_start, df['Actual'][zoom_start]), xytext=(zoom_start + 1, df['Actual'][zoom_start] + 50),
             )

plt.savefig('Graphicall_OUTPUT_Texas.pdf',bbox_inches='tight')
plt.savefig('Graphicall_OUTPUT_Texas.eps',bbox_inches='tight')
plt.savefig('Graphicall_OUTPUT_Texas.png',bbox_inches='tight')
# Show the plot
plt.show()
```
